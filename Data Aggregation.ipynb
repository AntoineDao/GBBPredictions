```python
import os
import pandas as pd
import re

bakers = []
episodes = []
rounds = []

for directory in os.listdir():
    if os.path.isdir(directory) and directory != '.ipynb_checkpoints' and directory != 'season_9' and directory != 'data':
        season_df = pd.read_csv('{}/episodes.csv'.format(directory))
        bakers_df = pd.read_csv('{}/bakers.csv'.format(directory))
        elimination_df = pd.read_csv('{}/elimination.csv'.format(directory))
        
        bakers.append(bakers_df)
        episodes.append(season_df)
        rounds.append(elimination_df)
        
bakers_df = pd.concat(bakers)
episodes_df = pd.concat(episodes)
rounds_df = pd.concat(rounds)
```

```python
regex = re.compile(r'\d')

columns_to_drop = [col for col in episodes_df.columns if regex.match(col)]
episodes_df = episodes_df.drop(columns_to_drop, axis=1)

regex = " ?(.*)s?$"

def clean_string(string):
    string = string.replace('.29', 'e')
    string = string.replace('_', ' ')
    string = string.replace('.C3.A2', 'a')
    string = string.replace('.28', '')
    string = string.lower()
    string = re.findall(regex, string)[0]
    
    string = string.replace(' semi-finale', '')
    string = string.replace(' quarter-finale', '')
    string = string.replace(' quarter finale', '')
    string = string.replace(' quarterfinale', '')
    string = string.replace(' finale', '')
    string = string.replace('unconventional', 'alternative')
    
    if string == 'cakes':
        string = 'cake'
        
    if string == 'biscuits and traybakes':
        string = 'biscuits'
        
    if string == 'cake':
        string = 'cakes'
        
    if string == 'pies and tarts':
        string = 'pies'
    
    return string

def get_rank(string):
    string = str(string)
    string = re.sub(r"\D*",'', string)
    return string
    
    
# themes = episodes_df
# themes['episode_theme'] = themes['episode_theme'].apply(lambda x: clean_string(x))

# themes.groupby('episode_theme')['Baker'].count().reset_index().sort_values(by='Baker')

episodes_df['episode_theme'] = episodes_df['episode_theme'].apply(lambda x: clean_string(x))

episodes_df['Technical'] = episodes_df['Technical'].apply(lambda x: get_rank(x))

episodes_df
```

```python
columns_to_join = ['Baker [6][7]', 'Baker [6][7][8]', 'Baker[6]', 'Baker[3]', 'Baker']

bakers_df['baker'] = bakers_df[columns_to_join].fillna('').sum(axis=1)

bakers_df = bakers_df.drop(columns_to_join, axis=1)
bakers_df = bakers_df.drop(['links', 'Links'], axis=1)

regex = re.compile(r'\d')

columns_to_keep = [col for col in bakers_df.columns if regex.match(col)]

bakers_df.head()
```

```python
regex = re.compile(r'\d')

rounds_bakers = rounds_df['Baker'] 
rounds_season = rounds_df['season']

columns_to_keep = [col for col in rounds_df.columns if regex.match(col)]
rounds_df = rounds_df[columns_to_keep]

rounds_df['5'] = rounds_df[['5', '5[a]']].fillna('').sum(axis=1)
rounds_df['6'] = rounds_df[['6', '6[1]', '6[b]']].fillna('').sum(axis=1)
rounds_df['7'] = rounds_df[['7', '7[1]']].fillna('').sum(axis=1)

rounds_df = rounds_df.drop(['5[a]', '6[1]', '6[b]', '7[1]'], axis=1)

rounds_df['baker'] = rounds_bakers
rounds_df['season'] = rounds_season

value_vars = [col for col in rounds_df.columns if regex.match(col)]

rounds_df = pd.melt(rounds_df, id_vars=['baker','season'], value_vars=value_vars)
rounds_df.drop(['variable'], axis=1)

rounds_df.columns = ['baker', 'season', 'round', 'outcome']

rounds_df
```

```python
# fuzz is used to compare TWO strings
from fuzzywuzzy import fuzz

# process is used to compare a string to MULTIPLE other strings
from fuzzywuzzy import process

def assert_ratio(row, search_col, search_string, ratio):
    name = row[search_col]
    if fuzz.token_sort_ratio(name, search_string) > ratio:
        return True
    else:
        return False
```

```python
rounds_df['eliminated'] = rounds_df.apply(lambda row: assert_ratio(row, 'outcome', 'Baker was eliminated',85), axis=1)
rounds_df['star baker'] = rounds_df.apply(lambda row: assert_ratio(row, 'outcome', 'Baker was star baker',85), axis=1)
rounds_df['passed'] = rounds_df.apply(lambda row: assert_ratio(row, 'outcome', 'baker gor through to the next round',85), axis=1)
rounds_df['favourite'] = rounds_df.apply(lambda row: assert_ratio(row, 'outcome', "Baker was one of the judges' favourite bakers that week, but did not win star baker",85), axis=1)
rounds_df['least favourite'] = rounds_df.apply(lambda row: assert_ratio(row, 'outcome', "Baker was one of the judges' least favourite bakers that week, but was not eliminated.",85), axis=1)
rounds_df['runner up'] = rounds_df.apply(lambda row: assert_ratio(row, 'outcome', "Baker was a series runner-up.",85), axis=1)
rounds_df['winner'] = rounds_df.apply(lambda row: assert_ratio(row, 'outcome', "Baker was the series winner.",85), axis=1)


rounds_df
```

```python
directory = 'data'

if not os.path.exists(directory):
    os.makedirs(directory)

rounds_df.to_csv('{}/all_rounds.csv'.format(directory))
episodes_df.to_csv('{}/all_.episodes.csv'.format(directory))
```
